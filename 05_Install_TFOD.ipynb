## 4 Installing TensorFlow Object Detection API

Now we will install the API, during the installation we would also need to compile some protobuf files.

Protobuf or [Protocol buffers](https://developers.google.com/protocol-buffers/docs/pythontutorial) is a type of serialized structured data. They are a language-neutral, platform-neutral, format for serialzing data. It is more efficient than JSON, XML, pickle and text storage formats. 

Once you prepare a protobuf file, then using a special source code you can extract the information on a variety of platforms and programming languages. 

Here's a more [detailed explanation](https://developers.google.com/protocol-buffers/docs/pythontutorial) of protocol buffers with an example."

In case you want to install **TFOD API** locally, you first need to install a **protobuf compiler** using:

`apt-get install protobuf-compiler`

However, Google Colab already has protobuf pre-installed and hence we need not require to install it manually in Colab.

**Note:** The `%%bash` command works only in Colab notebooks; and if you intend to install the API locally you should remove the `%%bash` command and run the remaining commands separately!

```python
# The `%%bash` magic command inside a notebook lets you run a cell run like a shell interface
# Note: the `bash` command works only on Colab.
%%bash 

# Change the directory to models/research
cd models/research/

# Compile the API's Protobuf files
protoc object_detection/protos/*.proto --python_out=.

# Copy the required Setup file
cp object_detection/packages/tf2/setup.py .

# Install the API using the setup.py file
python -m pip install .
```
